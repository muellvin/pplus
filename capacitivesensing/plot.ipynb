```python
import numpy as np
import matplotlib.pyplot as plt

# read the file
data = np.load('numerical_values_r_d_s.npy')

"""
data[i,j,k] corresponds to R1values[i], dvalues[j], svalues[k]
"""

"""
dvalues = np.array([2, 5, 10])*1e-3
R1values = np.array([100, 200, 300, 400, 500, 600, 700])*mil
svalues = np.array([10, 30, 50, 100, 200, 400, 600])*mil
"""

# create a 3D plot data[:,0,:]
mil = 0.0254e-3


# in m
R1values = np.array([50])*mil
dvalues = np.array([0.5, 1, 2, 3, 5, 10])*1e-3
svalues = np.array([10, 30, 50, 100, 200, 400])*mil

# convert to mm
R1values = R1values*1e3
dvalues = dvalues*1e3
svalues = svalues*1e3

subarray = data[0, :, :]

# Create a 3D plot
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Generate X, Y, Z coordinates for each point in the subarray
# X is the layer number, Y is the row number, and Z is the data value
X, Y = np.meshgrid(dvalues, svalues)
Z = subarray.T  # Transpose to match the dimensions of X and Y

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Label the axes
ax.set_xlabel('grounded plate distance d [mm]')
ax.set_ylabel('ground plane spacing s [mm]')
ax.set_zlabel('Capacity [pF]')

# Set the title
ax.set_title('Capacity as a function of d and s for R1 = 50mm')

# change the viewing angle
ax.view_init(30, 30)

ax.set_box_aspect(aspect=None, zoom=0.9)


# Show the plot
plt.show()

```
